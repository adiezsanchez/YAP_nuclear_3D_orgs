```python
from pathlib import Path
from tqdm import tqdm
import pandas as pd
```

```python
directory_path = Path("./raw_data/")
images = []

# Iterate through the lsm files in the directory
for file_path in directory_path.glob("*.czi"):
    # Remove Control and Isotype stainings from the analysis
    if "Control" and "Isotype" not in file_path.stem:
        images.append(str(file_path))
    
len(images)
```

```python
# Prepare a list to collect metadata for all images
metadata_list = []

for image in tqdm(images):

    # Read path storing raw image and extract filename and filename with extension
    file_path = Path(image)
    filename = file_path.stem
    filename_with_ext = file_path.name

    # Extract experiment, mouse, treatment and replica ids
    parts = filename.split(" ")
    experiment_id = parts[0]
    mouse_id = parts[1]
    treatment_id = parts[2]
    replica_id = parts[-1]

    # Append metadata to the list
    metadata_list.append({
        "filename": filename,
        "filename_with_ext": filename_with_ext,
        "experiment_id": experiment_id,
        "mouse_id": mouse_id,
        "treatment_id": treatment_id,
        "replica_id": replica_id
    })

# Create a DataFrame from metadata_list
df = pd.DataFrame(metadata_list)

# Save DataFrame to CSV
df.to_csv("file_list.csv", index=False)

# Display the DataFrame
display(df)
```

```python
# Extract all columns except those containing 'filename' in their name
non_filename_columns = [col for col in df.columns if 'filename' not in col]

# Print unique values for each column (except filename columns)
for col in non_filename_columns:
    unique_vals = df[col].unique()
    print(f"Unique values in {col} column:", unique_vals)
```
